```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
from sklearn.metrics import *
```

```python
'''   와인 분류하기
1 Decision Tree 사용해 보기
2 Random Forest 사용해 보기
3 SVM 사용해 보기
4 SGD Classifier 사용해 보기
5 Logistic Regression 사용해 보기 ''' 
```

```python
# (2) 데이터 준비, (3) 데이터 이해하기
wine = load_wine()
wine_data = wine.data
wine_label = wine.target
wine_df = pd.DataFrame(data=wine_data, columns=wine.feature_names)
wine_df
```

```python
wine_df["label"] = wine.target
wine_df
```

```python
print(wine_data.shape)
wine.target_names
```

```python

wine_df["label"].value_counts() 
```

```python
# (4) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier

# (5) 모델 학습 및 예측 
# (5) - 1 Decision Tree 
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# (5) - 2 Random Forest
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn import svm

svm_model = svm.SVC(random_state=32)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier(random_state=32)
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred, zero_division = 0))
#label2의 경우는 전혀 positive로 감지하지 않아 0값이 나오면서 오류가 발생한 부분을 해결하였다.
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(random_state=32, solver = 'liblinear')
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

## 와인 데이터 평가 지표

- 잘못된 제품으로 판별하면 안되기 때문에 FP와 FN 값을 줄여야 하며, 이 때 중요한 평가 지표로는 precision과 recall이 있다. 두 지표 값을 모두 고려할 수 있는 f1-score를 평가 지표로 사용해야한다.
- 데이터가 label2 표본이 적은 편인데, 대략적으로 48/178 label2가 아닐 확률이 73%나 된다. SGD Classifier 모델의 경우 2번을 아예 고르지도 않아서 precision, recall, f1score 모두 0이 나왔는데 이런 오차 속에서도 정확도가 0.67이나 나온다.

결론은 RandomForestClassifier 또는 DecisionTreeClassifier를 사용해야한다.
